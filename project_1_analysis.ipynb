### 2. **Introducing the Data**
The dataset contains three key columns:
- **Country**: The name of the country.
- **Code**: A three-letter country code representing each nation.
- **Population**: The population size of each country.
- **GDP per Capita**: The Gross Domestic Product per capita (a measure of economic prosperity per person) for each country.

This dataset allows us to explore the economic standing of countries and their population sizes. We will use this data to analyze patterns and relationships between economic performance and population.

```python
# Load the dataset
import pandas as pd
data = pd.read_csv('dictionary.csv')
data
```

### 3. **Pre-processing the Data**
Data preprocessing is crucial to ensure that the analysis is based on clean, reliable information. The preprocessing steps I followed include:
1. **Handling Missing Data**: Several countries have missing values in the GDP per capita column. I will decide how to handle these, either by filling missing values or removing those records.
2. **Data Cleaning**: Ensuring consistency in country names and codes by checking for any irregularities.
3. **Data Transformation**: Transforming columns if needed, such as scaling large values for better visualization.

```python
data.isnull().sum()
```

```python
# Drop rows where GDP per Capita is missing
cleaned_data = data.dropna(subset=['GDP per Capita'])

# Fill missing Population values with the median
cleaned_data['Population'].fillna(cleaned_data['Population'].median(), inplace=True)

# Confirm the cleaning process
cleaned_data.isnull().sum()
```

### 4. **Data Understanding and Visualization**
We will now visualize:
- The distribution of GDP per capita
- The relationship between population size and GDP per capita

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distribution of GDP per Capita
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data['GDP per Capita'], kde=True, bins=30)
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()

# Plot the relationship between Population and GDP per Capita
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='GDP per Capita', data=cleaned_data)
plt.title('Population vs GDP per Capita')
plt.xlabel('Population')
plt.ylabel('GDP per Capita')
plt.xscale('log')  # Using log scale for population to handle large differences
plt.show()
```

### 5. **Storytelling**
- The **distribution of GDP per Capita** shows a skewed distribution, with most countries clustering at lower GDP per capita levels and a few countries with much higher values. This suggests global economic inequality, where a few countries dominate in terms of wealth.
- The **Population vs GDP per Capita** shows no strong linear relationship between population size and economic prosperity. Some highly populated countries have low GDP per capita, while smaller nations may have higher GDP per capita.

From the analysis, we can see that the world is economically divided. Countries with smaller populations, like Luxembourg, tend to have higher GDP per capita, while larger nations such as India or China have significantly lower GDP per capita despite their massive populations. This suggests that while larger populations can drive industrial and economic output, it does not always translate into individual wealth.
